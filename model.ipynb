```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("data.csv")
```

```python
# Explorative Data Analysis
```

```python
sns.heatmap(data.isnull())
```

```python
data.drop(["Unnamed: 32", "id"], axis= 1, inplace=True)
```

```python
data.head()
```

```python
data.diagnosis = [ 1 if value == "M" else 0 for value in data.diagnosis ]
```

```python
data.head()
```

```python
data["diagnosis"].value_counts().plot(kind = "bar")
```

```python
# Predictors and Target Values
```

```python
y = data["diagnosis"] # Target
X = data.drop(["diagnosis"], axis = 1) #Predictor
```

```python
X.shape, y.shape
```

```python
X.head()
```

```python
# Normalisation of Data (X)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test  =train_test_split(X_scaled,y, test_size=0.30, random_state= 42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()

model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
y_predict
```

```python
y_test
```

```python
# Evaluation
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test , y_predict)
accuracy
```

```python
print(f"Accuracy: {accuracy: .2f}")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict))
```

```python
# Graph and Confusion Matrix
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt
```

```python
y_probs = model.predict_proba(X_test)[:, 1]

fpr, tpr, thresholds = roc_curve(y_test, y_probs) # False Positive Rate, True Positive Rate, thresholds

plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc_score(y_test, y_probs):.2f})")
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay


y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```
